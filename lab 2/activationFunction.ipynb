# 1 Binary Step Activation function

```python
def binaryStep(x):
        return np.heaviside(x,1)
```

```python
x=np.linspace(-10,10)
plt.plot(x,binaryStep(x))
plt.title("Binary Step")
plt.show()
```

# 2 Linear Activation function

```python
def linearAct(x):
        return x
```

```python
x=np.linspace(-10,10)
plt.plot(x,linearAct(x))
plt.title('Linear Activation')
plt.show()
```

# 3 Sigmoid Activation

```python
def sigmoidAct(x):
        return 1/(1+np.exp(-x))
```

```python
x=np.linspace(-10,10)
plt.plot(x,sigmoidAct(x))
plt.title('Sigmoid Activation')
plt.show()
```

# 4 TanH Activation

```python
def tanhAct(x):
        return np.tanh(x)
```

```python
x=np.linspace(-10,10)
plt.plot(x,tanhAct(x))
plt.title('tanh Activation')
plt.show()
```

# 5 Rectified Linear Unit (ReLU) Activation

```python
def reluAct(x):
    x1 = []
    for i in x:
        if i < 0:
            x1.append(0)
        else:
            x1.append(i)
    return x1
```

```python
x=np.linspace(-10,10)
plt.plot(x,reluAct(x))
plt.title('RELU Activation')
plt.show()
```

# 6 Softmax Activation

```python
def softmaxAct(x):
        return np.exp(x)/np.sum(np.exp(x))
```

```python
x=np.linspace(-10,10)
plt.plot(x,softmaxAct(x))
plt.title('Soft Max')
plt.show()
```
